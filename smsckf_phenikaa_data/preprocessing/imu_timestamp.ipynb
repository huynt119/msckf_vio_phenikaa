```python
def extract_timestamps_to_txt(df, start_time, output_file, duration):
    """
    Lọc các timestamp trong khoảng [start_time, end_time] và ghi vào file .txt

    Args:
        df (pd.DataFrame): DataFrame chứa dữ liệu
        timestamp_col (str): Tên cột timestamp
        start_time (float): Thời gian bắt đầu (epoch hoặc float)
        end_time (float): Thời gian kết thúc
        output_file (str): Đường dẫn file .txt để ghi
    """

    # Ép kiểu timestamp về float nếu cần
    timestamps = df['timestamp'].astype(float)
    end_time = start_time + duration * 1e3

    # Lọc khoảng thời gian
    mask = (timestamps >= start_time) & (timestamps <= end_time)
    filtered_timestamps = timestamps[mask]

    # Ghi ra file
    with open(output_file, 'w') as f:
        for ts in filtered_timestamps:
            f.write(f"{ts:.0f}\n")  # có thể đổi format nếu cần

    print(f"Đã ghi {len(filtered_timestamps)} dòng vào {output_file}")


def extract_timestamps_to_csv(df, start_time, output_file, duration):
    """
    Lọc các dòng có timestamp trong khoảng [start_time, start_time + duration]
    và ghi vào file .csv

    Args:
        df (pd.DataFrame): DataFrame chứa dữ liệu (phải có cột 'timestamp')
        start_time (float): Thời gian bắt đầu (epoch time hoặc float, đơn vị ms)
        output_file (str): Đường dẫn file .csv để ghi
        duration (float): Khoảng thời gian (đơn vị giây)
    """

    # Tính thời gian kết thúc
    end_time = start_time + duration * 1000  # chuyển từ giây sang ms

    # Ép kiểu timestamp về float nếu cần
    df['timestamp'] = df['timestamp'].astype(float)

    # Lọc các dòng theo khoảng thời gian
    mask = (df['timestamp'] >= start_time) & (df['timestamp'] <= end_time)
    filtered_df = df[mask]

    # Ghi ra file CSV
    filtered_df.to_csv(output_file, index=False)

    print(f"Đã ghi {len(filtered_df)} dòng vào {output_file}")
```
