```python
import pandas as pd
import folium

```

```python
df = pd.read_csv('./data_raw/ground_truth.csv')
df
```

```python
df = df[df['timestamp'] < 1743694270800]
df
```

```python

m = folium.Map(location=[df[' lat'][0], df[' lon'][0]], zoom_start=15)

# for _, row in df.iterrows():
#     folium.Marker([row[' lat'], row[' lon']], popup=row['timestamp']).add_to(m)

coordinates = df[[' lat', ' lon']].values.tolist()
folium.PolyLine(locations=coordinates, color='blue').add_to(m)

m.save("multi_points_map.html")
```
